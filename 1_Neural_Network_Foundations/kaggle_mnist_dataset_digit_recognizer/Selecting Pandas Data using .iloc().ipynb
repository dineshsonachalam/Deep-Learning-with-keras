```python
import pandas as pd
data = pd.read_csv("2010_Census_Populations_by_Zip_Code.csv")
data
```

```python
# Selection of all rows && from 2nd column till last column
data.iloc[:,1:]
```

```python
# Use.values to return values as a np array
data.iloc[:,1:].values
```

```python
# To get 1st row of the data frame
data.iloc[0]
```

```python
# To get .values of 1st rows i.e., return values as np array
data.iloc[0].values
```

```python
# TO get 2nd row of the data frame
data.iloc[1]
```

```python
# To get values of 2nd row of data frame as np array
data.iloc[1].values
```
